```python
from sklearn.experimental import enable_iterative_imputer  # Required to enable IterativeImputer
from sklearn.impute import IterativeImputer
from sklearn.ensemble import HistGradientBoostingRegressor

# Columns to impute
columns_to_impute = [
    "booked_days",
    "occupancy_rate",
    "booked_days_avePrice",
    "revenue",
    "prev_revenue",
    "prev_booked_days_avePrice",
    "prev_occupancy_rate",
    "prev_booked_days",
    "Cleaning Fee (USD)",
    "prev_prop_5_StarReviews_pastYear",
    "prev_rating_ave_pastYear",
    "prev_num_5_star_Rev_pastYear",
    "prev_numCancel_pastYear",
    "prev_numReviews_pastYear",
    "available_days_aveListedPrice_tractQuartile",
    "prev_Rating Overall",
    "available_days_aveListedPrice",
    "available_days",
    "prev_numReserv_pastYear",
    "prev_numReservedDays_pastYear",
    "prev_available_days_aveListedPrice_tractQuartile",
    "prev_available_days_aveListedPrice",
    "prev_available_days",
    "prev_Nightly Rate_tractQuartile",
    "prev_Nightly Rate",
    "prev_Number of Reviews"
]

# Initialize IterativeImputer with HistGradientBoostingRegressor
imputer = IterativeImputer(estimator=HistGradientBoostingRegressor(), max_iter=10, random_state=0)

# Apply imputation only on the selected columns
df_imputed = df.copy()
df_imputed[columns_to_impute] = imputer.fit_transform(df_imputed[columns_to_impute])
```
